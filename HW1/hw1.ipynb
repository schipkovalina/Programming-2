```python
# Программа предлагает пользователю выбрать одну из трех тем
# и загружает список слов, соответствующий выбору
def theme():
    i = input('Выберите одну тему и введите ее номер: 1) Хобби 2) Виды спорта 3) Экзотические фрукты \n ')
    while True:
        if i == "1" or i == '2' or i == '3':
            with open(i + '.txt', encoding='utf-8') as f:
                f = f.read()
                f = f.split("\n")
            break
        else:
            i = input('Введите только номер: 1) Хобби 2) Виды спорта 3) Экзотические фрукты \n ')
    return f
```

```python
# Выбор случайного слова из списка
def сhoice(f):
    word = random.choice(f)
    return word
```

```python
# Обработка ввода пользователя
letters = []
def сheck():
    while True:
        attempt = input('Какая буква может быть в этом слове? ').lower()
        letters.append(attempt)
        if attempt in letters[:-1]:
            print('Эту букву Вы уже пробовали')
        elif len(attempt) == 0:
            print('Вы ничего не ввели')
        elif attempt not in 'ёйцукенгшщзхъфывапролджэячсмитьбю':
            print('Введите букву из русского алфавита')
        else:
            return attempt
```

```python
# Программа выводит нижние подчеркивания вместо неотгаданных букв
# Если буква, прошедшая обработку ввода, есть в слове, то она открывается
# Выводится число попыток,максимальное - 10
# За попытку не считается некорректный символ и буква, которая уже была

def game(word):
    warning = 'У вас есть 10 попыток, чтобы угадать слово из ' + str(len(word)) + ' букв'
    print(warning)
    underscore = len(word) * '_ '
    print(underscore)
    attempts = 0
    while attempts < 10 and ''.join(underscore.split()) != word:
        guess = сheck()
        if guess in word:
            for index in range(len(word)):
                if word[index] == guess:
                    underscore = underscore[:index*2] + word[index] + ' ' + underscore[(index+1)*2:]
            print(underscore)
        else:
            print('В этом слове нет такой буквы!')
            attempts += 1
            print(draw()[attempts-1])
            if attempts == 9:
                print('Осталась 1 попытка')
            elif attempts >= 6 and attempts <= 8:
                print('Осталось', 10 - attempts, 'попытки')    
            else:
                print('Осталось', 10 - attempts, 'попыток')
            print(underscore)    
    return underscore, word
```

```python
# Сравнение угаданного с загаданным словом
def result(game):
    if ''.join(game[0].split()) == game[1]:
        print('Вы выиграли!')
    else:
        print('Вы проиграли :(')
    again()
```

```python
# Программа предлагает начать игру заново
def again():
    letters.clear()
    ii = input('Хотите начать заново? Введите "да" или "нет": ')
    if ii.lower() == "да":
        result(game(сhoice(theme())))
    if ii.lower() == "нет":
        print('Спасибо за игру!')
```
